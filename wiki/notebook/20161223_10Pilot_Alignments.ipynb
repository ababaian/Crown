## Pilot Pilot Run

To set-up that analysis; I'm going to have to run a pilot of the pilot run. To make this infomrative and quick, I'll use a low-coverage WGS data-set which should show the difference between high coverage and low-coverage WGS over rDNA.

````bash
# EC2 Command History

# Make working directory
    mkdir align; cd align

# Download Genoem Sequencing Data
  wget ftp://ftp.sra.ebi.ac.uk/vol1/fastq/SRR794/SRR794330/SRR794330_1.fastq.gz
  wget ftp://ftp.sra.ebi.ac.uk/vol1/fastq/SRR794/SRR794330/SRR794330_2.fastq.gz

# Copy hgr genome and create bowtie2 index
  cp ~/resources/hgr.fa ./
  cp ~/resources/hgr.fa.fai ./
  
  bowtie2-build hgr.fa hgr
  
# Read Group Data
RGID='SRR794330' # Read Group ID. Accession Number
RGSM='NA19240'   # Sample. Patient Identifer
RGLB='SRR794330' # Library Name. Accession Number
RGPL='ILLUMINA'  # Sequencing Platform.
# Extract Sequencing Run Info
RGPU=$(gzip -dc SRR794330_1.fastq.gz | head -n1 - | cut -f1 -d':' | cut -f2 -d' ')


# Bowtie2: align to genome
bowtie2 \
--very-sensitive-local -p 2 \
--rg-id $RGID --rg LB:$RGLB \
--rg SM:$RGSM --rg PL:$RGPL \
--rg PU:$RGPU \
-x hgr -1 SRR794330_1.fastq.gz -2 SRR794330_2.fastq.gz
| samtools view -bS - > aligned_unsorted.bam

bowtie2 --very-sensitive-local -p 2 --rg-id $RGID --rg LB:$RGLB --rg SM:$RGSM --rg PL:$RGPL --rg PU:$RGPU -x hgr -1 SRR794330_1.fastq.gz -2 SRR794330_2.fastq.gz -S aligned_reads.sam

# samtools view -bS aligned_reads.sam > aligned_unsorted.bam
# rm aligned_reads.sam

samtools sort aligned_unsorted.bam aligned
samtools index aligned.bam
rm aligned_unsorted.bam

# Re-opened in AM: changed to EC2 m4.xlarge instance

# Extract mapped reads, and their unmapped pairs
# surprisingly difficult...

# Unmapped reads with mapped pairs

    # Extract Mapped Reads
    # and their unmapped pairs
    samtools view -b -F 4 aligned.bam > align.F4.bam #mapped
    samtools view -b -f 4 -F 8 aligned.bam > align.f4F8.bam #unmapped pairs
    
    # Read Counts: align.F4 and align.f4F8
    # 4216876 Reads: 2098501 read1  2118375 read2
    # 1045242 Reads:  532558 read1   512684 read2
    # 5262118        2631059 read1  2631059 read2    
    
    # Extract Header
    samtools view -H align.F4.bam > align.header.tmp
        
    # Extract just the 45S unit
    echo -e "chr13\t999000\t1013500" > rDNA_45s.bed
    samtools view -b -L rDNA-45s.bed align.F4.bam > align.F4.45s.bam
    
    # What are the 45S mapped readnames
    samtools view align.F4.45s.bam | cut -f1 - > read.names.45s.tmp
    
    # Extract cases of read pairs mapped on edge of cut-off
    samtools view align.F4.bam | grep -Ff read.names.45s.txt - > align.F4.tmp.sam

    # Complete mapped reads list
    cut -f1 align.F4.tmp.sam > read.names.45s.long.tmp

    # Extract unmapped reads with a mapped pair in 45s.long
    samtools view align.f4F8.bam | grep -Ff read.names.45s.long.tmp - > align.f4F8.tmp.sam

    # Re-compile bam file
    cat align.header.tmp align.F4.tmp.sam align.f4F8.tmp.sam | samtools view -bS - > align.45s.tmp.bam
      samtools sort align.45s.tmp.bam align.45s
      samtools index align.45s.bam
      rm align.45s.tmp.bam
      
    # Read Counts: align.45s
    # 66736 Reads
    # 60877 mapped (91.22%)
    # 33368 read1; 33368 read2
    
    # Recompile the total rDNA mapped read pairs
    samtools cat -h align.header.tmp -o align.rDNA.bam align.F4.bam align.f4F8.bam
    samtools index align.rDNA.bam
    
    
    rm *tmp* # Clean-up

# Rename the final Bam Files
    mv aligned.bam NA19240_lowcov.bam
    mv aligned.bam.bai NA19240_lowcov.bam.bai

    mv align.rDNA.bam NA19240_lowcov.rDNA.bam
    mv align.rDNA.bam.bai NA19240_lowcov.rDNA.bam.bai
    
    mv align.45s.bam NA19240_lowcov.45s.bam
    mv align.45s.bam.bai NA19240_lowcov.45s.bam.bai

# aws s3 cp aligned.bam s3://crownproject/1kg_pilot/NA19240_lowcov.bam
# aws s3 cp aligned.bam.bai s3://crownproject/1kg_pilot/NA19240_lowcov.bam.bai
# Closed instance for the night

# GATK variant calling over 45S region
  aws s3 cp s3://crownproject/resources/hgr.gatk.fa ./
  aws s3 cp s3://crownproject/resources/hgr.gatk.fa.fai ./
  aws s3 cp s3://crownproject/resources/hgr.gatk.dict ./
  
  java -Xmx12G -jar /home/ubuntu/software/GenomeAnalysisTK.jar \
  -R hgr.gatk.fa -T HaplotypeCaller \
  -ploidy 2 --max_alternate_alleles 6 \
  -I NA19240_lowcov.bam -o NA19240.rDNA_p2.vcf
   # Memory issues, restrict to 45S region only
       # -ploidy 100, 50, 20 failed... do 2 and analyze 45S further
       
  java -Xmx12G -jar /home/ubuntu/software/GenomeAnalysisTK.jar \
  -R hgr.gatk.fa -T HaplotypeCaller \
  -ploidy 50 --max_alternate_alleles 10 \
  -I NA19240_lowcov.45s.bam -o NA19240.45s_p50.vcf


# Upload full file to amazon Glacier valut
aws s3 cp s3://crownproject/scripts/treehash.py ./
 chmod 755 treehash.py
aws s3 cp s3://crownproject/scripts/glacierupload.sh ./
 chmod 755 glacierupload.sh
 
 ./glacierupload.sh NA19240_lowcov.bam $AWSID
 
 aws s3 cp NA19240_lowcov.bam.glacer s3://crownproject/1kg_pilot/


# Upload final output files to S3
 
# 45S Only Alignent
 aws s3 cp NA19240_lowcov.45s.bam s3://crownproject/1kg_pilot/
 aws s3 cp NA19240_lowcov.45s.bam.bai s3://crownproject/1kg_pilot/
 
# rDNA Only Alignent
 aws s3 cp NA19240_lowcov.rDNA.bam s3://crownproject/1kg_pilot/
 aws s3 cp NA19240_lowcov.rDNA.bam.bai s3://crownproject/1kg_pilot/

#VCF over 45S
 aws s3 cp NA19240.45s_p50.vcf s3://crownproject/1kg_pilot/
 aws s3 cp NA19240.45s_p50.vcf.idx s3://crownproject/1kg_pilot/
 
#VCF over rDNA
 aws s3 cp NA19240.rDNA_p2.vcf s3://crownproject/1kg_pilot/
 aws s3 cp NA19240.rDNA_p2.vcf.idx s3://crownproject/1kg_pilot/

  
 # samtools view file.bam | grep queryname - > subset.sam
````

### Variation in the 45s rDNA components

![28S Pilot Genomes Variation](../figure/20170114_28S_pilotVariation.png)

Running the pilot genomes makes it immediatly apparant that the variations in each genome are based on some sort of biological/chemical truth. Sequencing more people in each population and looking for clustering may be an important finding, each population having it's own private pool of rDNA variation.

#### 28S A 59 A/G

One of the most interesting alleles is 28S.A59. In every genome examined, this nucleotide is maintained at near 50% allele frequency of A and G. Unlike most other variations, there is little variability in alternative allele frequency, and it's maintained. The evolutionary entropy at that position is also 0.565, meaning it's conserved.

This is a very good indicator of a functional consequence to this variation; it's adjacent to L15 and L29 binding sites.
