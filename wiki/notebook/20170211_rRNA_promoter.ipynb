## Introduction

The standard rDNA array reference begins with the start of the 47S pre-rRNA subunit which means the core upstream promoter region was located near positions ~43,000 (the far end). This region is obviously biological pertinant to core rRNA transcription/regulation and is located at an inconvinient place. For refinement of the rDNA refeference sequence the ~1 kb promoter region of rDNA will be  cut/pasted from the end of the reference sequence and into the start such that 47S will start at position 1,000 of hgr sequence.

Immediatly from the NA19240 alignment there is also significant variation between the aligned reads and the reference sequence so this will be refined. Note: [Zhou et al.,](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC5047480/) also noted sequence variation in the core promoter region (~200 bp). I'm using their definition of +1 TSS.

### Alternative Promoter Haplotypes

There isn't any evidence with the given sequence of a high-abundance alternative promoter haplotype in rDNA. The sequences seem to go into one main promoter sequence, the refined reference provided above.

![Promoter Haplotype](../figure/20170211_promoter_haplotype.png)

That horizontal green (A) reads present at ~2 reads were analyzed; representative read was Blat'ed to hg38 and the sequence matches an AluSz at chr1:156,216,371-156,216,496.
