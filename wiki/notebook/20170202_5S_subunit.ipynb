## Introduction

Analysis to this point has focused on 45S sequence which matures to the 18S, 5.8S and 28S rRNA


The final RNA species which is part of the ribosome is the 5S rRNA. 5S is transcribed from a distinct locus on chromosome 1. The 5S locus is also in tandem array, with a copy number ~equal to 45S array copy number.

In the human genome the 5S locus is at 1q42.13.

![1q42.13](../figure/20170206_1q42.13.png)

Zoomed into a single repeat

![5S repeat](../figure/20170206_5s_repeat_unit.png)


There is already some dbSNP data for 5S repeats. In the edge-repeat 18 there is a SNP within 5S [rs102046644](http://www.ncbi.nlm.nih.gov/SNP/snp_ref.cgi?type=rs&rs=rs12046644).

![dbSNP 147](../figure/20170206_5SP18_snp.png)

It's a logical extension to include this locus in rRNA variation analysis (namely cancer-specific variations) and to this end I'll be looking at how to best include a reference sequence.

## Objective

- Develop a 5S rRNA repeat reference sequence for alignment/variation analysis in 1kg project data.

## Alu Insertion in 5S Array

I've been thinking about how to deal with this Alu element. For the first iteration, of the analysis it's probably best to mask the Alu to N's and focus on sequence variation in the 5S itself.

![From Talk](../figure/20170213_5s_Alu.png)

- There is a high proportion of pseudo-5S rDNA which maps to this locus when no genomic sequence is avaialble.
- The better matching 'true 5s' sequence which is in the context of the 5S-repeat (flanking sequence matches) can also be non-RT pseudo 5S arrays but it's worth noting that all of those sequences in this low coverage experiment flank into the same Alu sequence, labeled 'true 5S'.
- It appears that all 5S repeats contain the same Alu insertion.

## Reference 5S array

For next round of alignments, I will include a single 5S repeat with AluY element masked. Reasoning here that sequence variation in the 5S itself will be the 'interesting bit'. If the above alignment is any indication this will be much messier, but if 'concordant' read pairs are looked at then the sequence analysis should clean up quite significantly.
