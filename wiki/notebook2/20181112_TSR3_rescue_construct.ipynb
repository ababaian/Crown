#### TSR3 Codon Optimized

The native TSR3 mRNA contains regions of highly skewed GC-content and secondary structure, ran through the [IDT codon optimization web tool](https://www.idtdna.com/CodonOpt). Settings: `Gene`, and `Homo Sapiens`.

Note there is a hairpin in the 5' UTR which ends at the M_1 AUG start codon, this is retained in the synthetic construct.
