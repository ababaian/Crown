#### 170403 -- Growing HT115 Cells

[HT115](https://www.phe-culturecollections.org.uk/products/celllines/generalcell/detail.jsp?refId=85061104&collection=ecacc_gc) cells (cat# ECACC 85061104) were taken from FLs box; 293I (oct 3rd). Unknown passage, unknown starting density.

Cells initially are clumpy/round (first thaw) but flatten out and look like typical fibroblasts. Stated as growing in 'Islands' but I'm not sure I see this, they are perhaps a bit patchy. ~5% of the cells round up under these growth conditions and shed into suspension, possibly worth changing media between passages depending on how they look.

##### Growth Medium
In hood mixed up some growth medium.
- DMEM 500 mL (Stemcell)
- 75 mL FBS (15%)
- 5.75 200 mM L-glutamine (2 mM) (Stemcell)

##### Thawing Cells
1 mL of cells in the freezer media were directly plated into 5mL of *4'C cold* media (5 mL in T25).

Next AM media exchanged after 1x PBS wash. Arbitrarily designated cells passage 10.

##### Growing Cells
Passage cells at between 70-80% confluency

Cells should be seeded at 2-4 e4 cells/ml^2.
	T25: 5e5-1e6
	T75: 2e6-4e6

#### 170405
Cells are at ~80% confluency. Cells passaged from T25 flask and seeded into a T75 flask.

##### Passaging Cells
- Old media is removed, cells are washed in 5 mL warmed PBS
- 1.5 mL (T25), 2 mL (T75) of Trypsin-EDTA is added to the cells, incubated at 37'C for 1-3 minutes untill cells dislodge with tapping.
- 3.5 mL (T25) or 8 mL (T75) growth media added to inactive Trypsin.
- Cells spun at 300xg for 5 minutes to pellet
- Pellet resuspended in 5 mL (T25) or 10 mL (T75) PBS for wash and spun again
- When counting cells, 0.5 mL of cell suspsenion aliquoted here for further dilution/counting.
- Cells resuspended in 5 mL (T25) or 10 mL (T75) media and seeded at the appropriate density.

##### Cell Counting
Unless otherwise noted this is the protocol for counting cells; I often simply report the two cell count numbers and the final concentration.

- Aliquot of cells is resuspended / mixed to homogenize. Ensure a fine single-cell suspension here (no clumps)
- 10 uL of cell suspension is mixed with 90 uL of Trypan-Blue for a 10x dilution
- 10 uL of the diluted cells are plated into a Hemocytometer, both sides independently.
- Count the 5 squares, corners and middle. Top and Right boundary is considered 'in'.
- 59 cells in first count; 43 cells in second count.
- Sum both counts for a count in 10 squares total. (102 in this case)
- 102 cells in 1 mm^3 or in 1 uL

- <Cells> * <Dilution Factor> * 1000 (for mL)
- 102 * 10 * 1000 = 1.02 x 10^6 cells/mL
- 1.02e6 cells/mL with a total of 5 mL suspension

#### HCT115 p13 - Trizol Extraction
A T75 flask of HT115 were grown to ~80% confluency. Cells were trypsinized by SOP and washed 1x in 10 mL PBS and counted.

- Cells were resuspended in PBS again and 3e6 cells in suspension were aliquoted into 1.5 mL tubes in triplicate
- Cells were pelleted at 500g 5m
- 0.5 mL trizol reagent was added to the 3e6 cell pellets
- Pellet was pipetted and vortexed to homogenize.
- Tubes incubated for 5 minutes at RT and then immediatly placed into -80'C freezer

* Frozen samples kept in -80'C DM Freezer #7, Rack 8, Box 1 'Artem HL Box'

#### RNA/DNA Extraction
Followed [Trizol](https://tools.thermofisher.com/content/sfs/manuals/trizol_reagent.pdf) SOP for RNA/DNA extraction with the following modifications:

- Two HCT115 Trizol samples thawed on ice, incubated 5 minutes at RT
- 0.5 mL of reagent added to 3e6 washed cell pellet. Volumes adjusted accordingly throughout.

* DNA WASH -- Sodium Citrate Solution (0.1 M)
- 50 mL ddH2O
- 1.4705 g Dihydrate sodium Cirate (294.1 g /mol)

* Sodium Hydroxide (1M)
- 100 mL ddH2O
- 3.9997 g NaOH
- warmed up at 37'C bath to dissolve

* DNA SUSPEND -- Sodium Hydroxide (8 mM)
- 49.6 mL ddH2O
- 0.4  mL 1M NaOH

- DNA resuspended in 100 uL 8mM NaOH ddH2O, stored in 4'C
