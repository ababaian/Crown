Notes:
- The issue with the original ADcalc script returning zeros is that the reads over 1248 are flagged 73/89/137 --> common setting of flag:8 Mate unmapped. Even though if you grep for a single read you can see above that both read and it's mate are in fact mapped eyt the 8 flag was still set. This is likely an issue with inserte size from teh aligner but can be corrected by adding the `-A` tag to `bcftools` runs.


````bash
#!/bin/bash
# ADcalc.sh
# Allelic Depth Calculator
# for a position

cd ~/tcga/

# Controls -----------------
REGION='chr13:1004908'
OUTPUT='18S_1248.vcf'
DEPTH='100000'
BAMLIST='bam.list.tmp'

# -A: do not discard anomolous read pairs

# Iterate through every TCGA Cancer Type
for TYPE in $(ls)
do
    echo Analyzing $TYPE...
    
    cd $TYPE

    ls *.bam > bam.list.tmp

    # Iterate through every bam file in directory
    # look-up position and return VCF
        bcftools mpileup -f ~/resources/hgr1/hgr1.fa \
      --max-depth $DEPTH --min-BQ 30 -A \
      -a FORMAT/DP,AD \
      -r "$REGION" \
      --ignore-RG \
      -b $BAMLIST |
      bcftools annotate -x INFO,FORMAT/PL - |
      bcftools view -O v -H - |\
      tail -n1 - \
      > ../$TYPE.$OUTPUT
      
    rm bam.list.tmp
    
    cd ..
done

# On AMI image with TCGA data loaded (crown-180906 ami-0ed79a4de8e2ca023)
# ran above script to generate:
# s3://crownproject/tcga/vcf_1248/
````
