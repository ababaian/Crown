## Introduction

From the CRC_1-3 experiments, I'm postulating that there are a sub-set of CRC which have hypo-modification of 18S macp-Phi 1248 and this may associate with some phonotype in those cells, such as permissive translation or antibiotic hypersensitivity.

The standard assays for measuring the hypermodification of macp1248 are; Primer Extension Assay and HPLC of individual bases. These are both a little bit restrictive for me to do a simple screen for but are good 'confirmatory' tools.

I'd like to be able to quickly assay cDNA and get a semi-quantitative measure for macp. In the RNA-seq how I identify the modification I beleive to be due to error-prone reverse transcriptase. This isn't dependent on a large sequencing run since the error is introduced at the RT step. So standard RT and PCR-based methods should be able to measure the same 40/40% T/C incorporation.

Thus the use of RT-PCR, sanger-based sequencing and/or Restriction Enzyme digestion will be the fastest way to assay a large amount of samples.



## Objective

* Develop a restriction enzyme-based assay to measure macp1248 in 18S at a given position

#### 170426 -- Error

In the lab as I was diluting these primers I noticed that both EcoRI primers have an extra base incorporated into them.
```
>Frag6_fwd1_EcoRI
TAGAT GAATTCC GTTCGAAGACGATCAGATACCG
>Frag6_fwd2_EcoRI
TAGAT GAATTCC TCAGATACCGTCGTAGTTCCG
```
should be
```
>Frag6_fwd1_EcoRI
TAGAT GAATTC GTTCGAAGACGATCAGATACCG
>Frag6_fwd2_EcoRI
TAGAT GAATTC TCAGATACCGTCGTAGTTCCG
```

It might not make a difference but I should order the correct primers as well.
