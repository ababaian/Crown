## Introduction

I have a hit at 18S.1248U for difference between cancer and normal. The *reference* allele (T) increases 11%, so this isn't an oncogenic mutation / variant allele frequency increase it's a difference in the post-transcriptional modification of this base to macp-Pseudouradine. I discuss this in the [manuscript](http://biorxiv.org/content/early/2017/03/21/118760) mainly as an interesting observation, low and behold that's the difference I see in cancer.

The assay for this position is Primer Extension of the rRNA, i.e. difficulty of reverse transcriptase in processing this position.

So let's get deep into this position a bit. It's perfectly conserved in Eukaryotes, the modification is the same in yeast and humans and E.coli/prokarya have another modification of the homologous position. This base sits at the core of the P-site.

The modification is [catalyzed](https://www.ncbi.nlm.nih.gov/pubmed/27084949) by two enzymes and a snoRNA; snoRNA 35 (Aca13 in humans) targets the base for pseudo-urydilation; Nep1 (EMG1 in human/mouse)  methyltransferase methylates N1; and finally Tsr3 adds the N3-aminocarboxylpropyl modification.

![macp](../../data/CRC/plot/macp_Psi.png)
From: Meyer et al., 2016

- Meyer et al., found that Yeast that are tsr3[knockout] (no macp) are viable but become **hypersensitive** to Paromomycin and partially hypersensitive to Hydromycin B.

Of course this is preliminary but if cancer's pheno-copy tsr3[ko] then potentially they become hypersensitive to Paromomycin or at least this may be a druggable target for which we have many starting molcules already. Pretty much ribosome-targetting antibiotics, especially things like Tetracyclin which target the P-site directly.

#### Mutations
Finally, a mutation in [EMG1](https://www.omim.org/entry/211180#1) causes the autosomal reccessive ribosomopathy Bowen-Conradi Syndrome. With the mutated allele frequency expected at 0.10 in the Canadian Hutterite anabaptist population. This may be highly applicable to this population as a line of chemotherapy. [Ref](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC3045603/) [Ref](https://www.ncbi.nlm.nih.gov/pubmed/19463982)

On the online database [tumorportal](http://dev.tumorportal.org/charts/gene), looking at TSR3 and EMG1 there isn't mutation of TSR3 but there appears to be a recurrent splice mutation in EMG1. Paradoxically, EMG1 nor any synonyms I could find for EMG1 (Nep1 etc...) is in the COSMIC database.
